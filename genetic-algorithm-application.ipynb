# **Coding of general functions and initial parameters**

> Genetik Algoritma kullanmak suretiyle verilen fonksiyonu **44** değerine eşitleyen **x1, x2, x3, x4, x5, x6** değerlerini tespit etmek istenmektedir.
> 
> **f(x1, x2, x3, x4, x5, x6) = 4x1 – 2x2 + 3,5x3 + 5x4 – 11x5 – 4,7x6**

> **pygad** kütüphanesi (genetik algoritma için) ve **numpy** kütüphanesi (sayısal işlemler için) ithal edilir.

> Genetik algoritmanın çözümlerinde kullanacağı **giriş değerleri** ve **hedef çıktısı** tanımlanır.

> **Uygunluk fonksiyonu**, çözümün ne kadar iyi olduğunu belirler. **solution** ve **function_inputs** çarpılarak toplamları hesaplanır ve bu toplamın **desired_output**'a olan uzaklığına göre uygunluk değeri hesaplanır. Uygunluk değeri, toplamın hedefe **ne kadar yakın** olduğuna bağlıdır.

> Genetik algoritmanın çalışması için gerekli olan bazı parametreler belirlenir:
> 
> * **num_generations:** Toplam nesil sayısı (iterasyon sayısı). Optimizasyon toplam döngü sayısı *100* nesil sayısı ile tanımlanmaktadır. 
> * **num_parents_mating:** Ebeveyn olarak seçilecek birey sayısı. Her döngüde *en iyi 7 birey* *çaprazlama, mutasyon* gibi işlemlerle yeni nesil popülasyon oluşturmak için seçilmektedir.
> * **sol_per_pop:** Popülasyon başına çözüm sayısı. Toplamda *50 adet* bireyden oluşan bir popülasyon söz konusudur.
> * **num_genes:** Her çözümdeki gen (değişken) sayısı.
> * **last_fitness:** Son neslin en iyi uygunluk değeri (başlangıçta sıfır).

> **nesil_ozeti** fonksiyonu, her nesil sonunda çağrılır ve nesil hakkında özet bilgi verir. *Nesil sayısı, en iyi çözümün uygunluk değeri* ve *bu değerin bir önceki nesilden olan değişimi* yazdırılır.

> Genetik algoritma parametrelerle birlikte oluşturulur ve çalıştırılır. **on_generation** parametresi ile her nesil sonunda **nesil_ozeti** fonksiyonu çağrılır. **run** fonksiyonu ile optimizasyon süreci başlatılır. 

> **En iyi çözüm (veya birey)**, **onun uygunluk değeri** ve **indeks numarası** elde edilir ve yazdırılır. Ayrıca, en iyi çözüm ile elde edilen **fonksiyon sonucu** hesaplanır ve yazdırılır. Eğer en iyi çözüm bir nesilde bulunmuşsa, bu **nesil sayısı** da yazdırılır.

> İlgili kodlara ek olarak **ga_instance.plot_fitness()** komutu yardımıyla en iyi/uygun değerin **değişim grafiği** de gösterilebilmektedir

```python
!pip install pygad

# Gerekli Kütüphanelerin İthal Edilmesi
import pygad
import numpy

# Girdi ve Hedef Değerin Tanımlanması
function_inputs = [4, -2, 3.5, 5, -11, -4.7] 
desired_output = 44 

# Uygunluk Fonksiyonunun Tanımlanması
def fitness_func(ga_instance, solution, solution_idx):
    output = numpy.sum(solution * function_inputs)
    fitness = 1.0 / numpy.abs(output - desired_output)
    return fitness

# Genetik Algoritmanın Parametrelerinin Ayarlanması
fitness_function = fitness_func

num_generations = 100 
num_parents_mating = 7

sol_per_pop = 50
num_genes = len(function_inputs)

last_fitness = 0

# Nesil Özeti Fonksiyonunun Tanımlanması
def nesil_ozeti(ga_instance):
    global last_fitness
    print("Nesil = {generation}".format(generation=ga_instance.generations_completed))
    print("Fonksiyon Sonucu = {fitness}".format(fitness=ga_instance.best_solution()[1]))
    print("Degisim = {change}".format(change=ga_instance.best_solution()[1] - last_fitness))
    last_fitness = ga_instance.best_solution()[1]

# Genetik Algoritmanın Oluşturulması ve Çalıştırılması
ga_instance = pygad.GA(num_generations=num_generations,
                       num_parents_mating=num_parents_mating, 
                       fitness_func=fitness_function,
                       sol_per_pop=sol_per_pop, 
                       num_genes=num_genes,
                       on_generation=nesil_ozeti)

ga_instance.run()

# En İyi Çözümün Elde Edilmesi ve Yazdırılması
solution, solution_fitness, solution_idx = ga_instance.best_solution()
print("En uygun cozum degisken degerleri : {solution}".format(solution=solution))
print("En uygun cozumu veren birey indeks no.: {solution_idx}".format(solution_idx=solution_idx))

prediction = numpy.sum(numpy.array(function_inputs) * solution)
print("En uygun cozum ile fonksiyon sonucu : {prediction}".format(prediction=prediction))

if ga_instance.best_solution_generation != -1:
    print("En uygun cozum {best_solution_generation} nesil sonra elde edildi."
          .format(best_solution_generation=ga_instance.best_solution_generation))

# Değişim grafiği çizdirme
ga_instance.plot_fitness()
```
